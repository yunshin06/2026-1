```python
#2490
def play(L):
    if L.count('0') == 0:
        print('E')
    elif L.count('0') == 1:
        print('A')
    elif L.count('0') == 2:
        print('B')

    elif L.count('0') == 3:
        print('C')
    elif L.count('0') == 4:
        print('D')
L_1 = input().split()
L_2 = input().split()
L_3 = input().split()
play(L_1)
play(L_2)
play(L_3)
```
