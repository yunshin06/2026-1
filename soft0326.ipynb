```python
# 10952
L = []
while True:
  A, B = map(int, input().split())
  if A == 0 and B == 0:
    break
  L.append(A+B)

for n in L:
  print(n)
```

```python
# 13410
X, Y = map(int, input().split())
L_1 = []
L_2 = []
for y in range(1, Y+1):
  xy = X*y
  L_1.append(xy)
for num in L_1:
  a = num // 10
  b = num % 10
  c = (b*10) + a
  L_2.append(c)
print(max(L_2))
```

```python
# 2446
num = int(input())
blank = 0
for n in range(2*num-1, 0, -2):
  print(' '*blank+'*'*n)
  blank = blank + 1
blank = num - 1
for n in range(3, 2*num, 2):
  blank = blank - 1
  print(' '*blank+'*'*n)
```

```python
#10996
num = int(input())
for n in range(num):
  if num == 1:
    print('*')
  elif num % 2 == 0:
    a = num//2
    print('* '*a)
    print(' *'*a)
  else:
    a = num//2+1
    b = num//2
    print('* '*a)
    print(' *'*b)
```
