```python
#2523
a = int(input())
for star in range(1, 2*a+1):
  if star <= a:
    print('*'*star)
  else:
    print('*'*((2*a)-star))
```

```python
#1110
cycle = 0
num = int(input())
n = num
while True:
    a = num // 10
    b = num % 10
    c = a + b
    num = (b * 10) + (c % 10)
    cycle = cycle + 1
    if num == n:
        break
print(cycle)
```
