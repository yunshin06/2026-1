```python
# 버블 정렬
a = list(map(int, input().split()))
for i in range(len(a)):
  for j in range(len(a)-1):
    if a[j] > a[j+1]:
      a[j], a[j+1] = a[j+1], a[j]
print(a)
```

```python
# 10816
N = int(input())
n = list(map(int, input().split()))
M = int(input())
m = list(map(int, input().split()))
dic = {}
for i in n:
  if i in dic:
    dic[i] += 1
  else:
    dic[i] = 1
for i in m:
  print(dic.get(i, 0), end = ' ')
```

```python
# 7785
S = set()
n = int(input())
for i in range(n):
  name, stay = input().split()
  if stay == 'enter':
    S.add(name)
  elif stay == 'leave':
    S.discard(name)
S_sort = sorted(S, reverse=True)
for i in S_sort:
  print(i)
```

```python
# 1764
n, m =input().split()
n = int(n)
m = int(m)
N = set()
M = set()
S = set()
for i in range(n):
  N.add(input().strip())
for i in range(m):
  M.add(input().strip())
S = N & M
S_sort = sorted(S)
print(len(S_sort))
for i in S_sort:
  print(i)
```

```python
# 1302
n = int(input())
sell = {}
for i in range(n):
  book = input()
  if book in sell:
    sell[book] += 1
  else:
    sell[book] = 1
result = sorted(sell.items(), key=lambda x: (-x[1], x[0]))
print(result[0][0])
```
