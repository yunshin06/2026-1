```python
month = int(input())
print(f'          {month}월')
print('일 월 화 수 목 금 토')
total = 0
a = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
for i in range(month-1):
  total = total + a[i]
b = (total + 4) % 7
print(' '*(3*b), end='')
c = 7 - b
for i in range(1, a[month-1] + 1):
  if i < 10:
    print(f' {i}', end=' ')
    if i % 7 == c:
      print()
  else:
    print(f'{i}', end=' ')
    if i % 7 == c:
      print()
```

```python
# 4344
num = int(input())
percent = []
for i in range(num):
  E = 0
  add = 0
  score = list(map(int, input().split()))
  n = score.pop(0)
  for j in score:
    E = E + j
  E = E / n
  for j in score:
    if j > E:
      add = add + 1
  per = (add / n) * 100
  percent.append(per)
for i in percent:
  print(f"{i:.3f}%")
```

```python
# 8958
n = int(input())
L = []
ad = 0
add = 0
for i in range(n):
  x = input().split('X')
  for j in x:
    for a in range(1, len(j) + 1):
       ad = ad + a
    add = add + ad
    ad = 0
  L.append(add)
  add = 0
for i in L:
  print(i)
```

```python
# 9012
n = int(input())
L = []
for j in range(n):
  x = input()
  a = 0
  b = 0
  for i in x:
    if i == '(':
      a = a + 1
    elif i == ')':
      b = b + 1
  if a != b:
    L.append('NO')
  elif x[0] == ')' or x[-1] == '(':
    L.append('NO')
  else:
    L.append('YES')
for i in L:
  print(i)
```
